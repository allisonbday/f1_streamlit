```python
import fastf1
from fastf1.ergast import Ergast

import pandas as pd
```

# Getting results in a season


```python
ergast = Ergast()

for i in range(2021, 2021):

    season = i
    races = ergast.get_race_schedule(season)  # [3:4]  # Races in year 2022
    results = []

    # For each race in the season
    for rnd, race in races["raceName"].items():

        # Get results. Note that we use the round no. + 1, because the round no.
        # starts from one (1) instead of zero (0)
        # temp = ergast.get_lap_times(season=2022, round=rnd + 1)

        session = fastf1.get_session(season, race, "R")
        session.load()

        # temp = session.laps # Get lap times
        temp = session.results  # Get results

        # Add round no. and grand prix name
        temp["round"] = rnd + 1
        temp["race"] = race.removesuffix(" Grand Prix")
        temp["sprint"] = "False"
        temp["season"] = season
        results.append(temp)

        # test
        try:
            session = fastf1.get_session(season, race, "S")
            session.load()
            # temp = session.laps
            temp = session.results  # Get results

            # Add round no. and grand prix name
            temp["round"] = rnd + 1
            temp["race"] = race.removesuffix(" Grand Prix")
            temp["sprint"] = "True"
            temp["season"] = season
            results.append(temp)

        except:
            pass

    # Append all races into a single dataframe
    results = pd.concat(results)
    races = results["race"].drop_duplicates()

    results.to_csv(f"../data/results/results_{season}.csv", index=False)
    
```

```python
def apply_pnts_dict(position, dict):
    if position in dict.keys():
        return dict[position]
    else:
        return 0
```

```python
def pnts_1962(season):
    season.groupby(["FullName"])["Position"].nlargest(5)
```

```python
laps_2023 = pd.read_csv("../data/lap_times/lap_times_2023.csv")

results_2023 = pd.read_csv("../data/results/results_2023.csv")

laps_2023 = laps_2023.merge(
    results_2023[["race", "DriverNumber", "sprint", "Status"]],
    on=["race", "DriverNumber", "sprint"],
).reset_index(drop=True)

laps_2023
```

```python
# 2024 points

race_pts_2010 = {
    1: 25,
    2: 18,
    3: 15,
    4: 12,
    5: 10,
    6: 8,
    7: 6,
    8: 4,
    9: 2,
    10: 1,
    11: 0,
    12: 0,
    13: 0,
    14: 0,
    15: 0,
    16: 0,
    17: 0,
    18: 0,
    19: 0,
    20: 0,
    21: 0,
    22: 0,
    23: 0,
}

fastest_lap_pnt_point = 1  # if in top 10

sprint_pts_2022 = {
    1: 8,
    2: 7,
    3: 6,
    4: 5,
    5: 4,
    6: 3,
    7: 2,
    8: 1,
    9: 0,
    10: 0,
    11: 0,
    12: 0,
    13: 0,
    14: 0,
    15: 0,
    16: 0,
    17: 0,
    18: 0,
    19: 0,
    20: 0,
    21: 0,
    22: 0,
    23: 0,
}


# 2024 points

# fastest lap
laps_2023["fastest_lap_pnt"] = (
    laps_2023.loc[laps_2023["Deleted"] == False]  # only valid laps
    .groupby("race")["LapTime"]
    .rank()
    .apply(lambda x: True if x == 1 else False)
)  # Fastest lap
laps_2023["fastest_lap_pnt"] = laps_2023.groupby(["Driver", "race", "sprint"])[
    "fastest_lap_pnt"
].transform("any")


# # 2024 points
results_2023 = (
    laps_2023.groupby(["Driver", "race", "sprint"])
    .last()
    .sort_values("round")
    .reset_index()
)

results_2023["points"] = 0

results_2023["points"] = results_2023.apply(
    lambda x: (
        sprint_pts_2022[x["Position"]]
        if x["sprint"] and x["Status"] != "Disqualified"
        elif x["Status"] != "Disqualified" race_pts_2010[x["Position"]]
        else 0
    ),
    axis=1,
)

# fastest lap point
results_2023["points"] = results_2023.apply(
    lambda x: (
        x["points"] + fastest_lap_pnt_point
        if (x["Position"] <= 10) & (x["fastest_lap_pnt"] == 1)
        else x["points"]
    ),
    axis=1,
)


results_2023
```
